## 4. Tempo de entrega

````python
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])
orders["order_delivered_customer_date"] = pd.to_datetime(orders["order_delivered_customer_date"])

orders["tempo_entrega_dias"] = (
    orders["order_delivered_customer_date"] - orders["order_purchase_timestamp"]
).dt.days

orders["tempo_entrega_dias"].describe()
````

**Insight:** o count (96.476) bate exatamente com o número de pedidos delivered (96.478, com diferença mínima por arredondamento/edge case). Confirma que o cálculo já está implicitamente restrito aos pedidos entregues, já que `dt.days` retorna `NaN` quando a data de entrega é nula.

A distribuição é bem concentrada: metade dos pedidos (mediana) chega em até 10 dias, e 75% chegam em até 15 dias. Só que a média (12,1) é puxada pra cima da mediana (10),  sinal de assimetria à direita, ou seja, existem pedidos que demoram muito mais que o normal e "puxam" a média.

O `max` de 209 dias é o outlier mais evidente, quase 7 meses pra entregar um pedido é um caso muito fora do padrão (desvio padrão de ~9,5 dias, então 209 está a mais de 20 desvios da média). Vale investigar se é:

- um erro de digitação/carga no dado original,
- um caso real de extravio recuperado tarde,
- ou um pedido com problema logístico grave (ex: endereço errado, greve, etc).

**Decisão:** antes de usar `tempo_entrega_dias` como métrica na camada Gold (ex: tempo médio de entrega por região), vale aplicar algum **tratamento de outlier** na Silver, seja removendo casos extremos (ex: > 60 dias) ou apenas sinalizando-os numa coluna separada, sem descartar o dado bruto.

## 5. Pagamentos

**Insight:** distribuição fortemente assimétrica, a mediana (R$ 100) é bem menor que a média (R$ 154), e o desvio padrão (217) é maior que a própria média, o que é um sinal claro de cauda longa (poucos valores muito altos distorcendo a média). O max de R$ 13.664,08 é um caso extremo, muito acima do 75º percentil (R$ 171,84).

O `min = 0.00` também chama atenção, um pagamento de valor zero é estranho (pode ser voucher, cupom, ou dado inconsistente) e merece checagem separada.

Como `payments` tem uma linha por parcela (não por pedido), os valores
podem precisar ser agregados por `order_id` antes de qualquer análise
de ticket médio.

**Decisão:** pra métricas de "ticket médio" na Gold, a mediana é uma medida mais robusta que a média nesse tipo de distribuição. Também vale investigar os pagamentos de R$ 0 e os valores muito altos antes de decidir se entram nas agregações ou são tratados à parte (ex: poderiam representar múltiplas parcelas somadas por pedido, já que a tabela payments tem uma linha por parcela, não por pedido).

Resumo: usar mediana como medida central em vez de média para "ticket médio" na Gold; investigar pagamentos de valor zero.

## 5. Investigação de outliers - tempo de entrega
1. Outlier de tempo de entrega 

````python
outlier_entrega = orders[orders["tempo_entrega_dias"] > 60].sort_values(
    "tempo_entrega_dias", ascending=False
)
outlier_entrega[[
    "order_id", "order_status", "order_purchase_timestamp",
    "order_delivered_customer_date", "order_estimated_delivery_date",
    "tempo_entrega_dias"
]].head(10)
````

````python
outlier_entrega.shape[0]
````

**Insight:** outliers de tempo de entrega

288 pedidos (cerca de 0,3% do total de entregues) levaram mais de 60 dias.

Parece ser um fechamento de sistemas em lote.

Datas de entrega no topo do ranking:

```text
2017-09-19 14:36:39
2017-09-19 15:12:50
2017-09-19 14:00:04
2017-09-19 14:33:17
2017-09-19 14:38:21
2017-09-19 17:00:07
...
2018-09-19 23:24:07
```

Não é caso isolado é um padrão.

Múltiplos pedidos completamente diferentes, comprados em datas distantes (fevereiro, março de 2017), todos aparecem como "entregues" no mesmíssimo dia: 19/09/2017 (e outro grupo em 19/09/2018, exatamente um ano depois).

Não acredito que o pedido  demorou 200 dias no caminhão. Pode ser que a Olist rodou algum job de "fechamento" ou reconciliação de pedidos travados/perdidos numa data específica, marcando todos como entregues de uma vez.

**Decisão:**
Vale checar se existe uma concentração de datas de entrega (`value_counts() em order_delivered_customer_date`) que confirme esse padrão em lote de forma mais ampla.

Excluir ou sinalizar esses registros como "entrega reconciliada em lote" antes de calcular métricas de SLA/tempo de entrega real na camada Gold, eles distorceriam a análise de performance logística.
